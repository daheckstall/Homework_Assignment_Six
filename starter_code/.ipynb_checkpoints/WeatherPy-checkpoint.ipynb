```python
# Dependencies and Setup
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
import requests
import time
import openweathermapy.core as owm

# Import API key
import api_keys

# Incorporated citipy to determine city based on latitude and longitude
from citipy import citipy

# Output File (CSV)
output_data_file = "output_data/cities.csv"

# Range of latitudes and longitudes
lat_range = (-90, 90)
lng_range = (-180, 180)
```

## Generate Cities List

```python
# List for holding lat_lngs and cities
lat_lngs = []
cities = []

# Create a set of random lat and lng combinations
lats = np.random.uniform(low=-90.000, high=90.000, size=1500)
lngs = np.random.uniform(low=-180.000, high=180.000, size=1500)
lat_lngs = zip(lats, lngs)

# Identify nearest city for each lat, lng combination
for lat_lng in lat_lngs:
    city = citipy.nearest_city(lat_lng[0], lat_lng[1]).city_name
    
    # If the city is unique, then add it to a our cities list
    if city not in cities:
        cities.append(city)

# Print the city count to confirm sufficient count
len(cities)
```

## Perform API Calls

```python
# OpenWeatherMap API Key
api_key = api_keys.api_key

# Starting URL for Weather Map API Call
url = "http://api.openweathermap.org/data/2.5/weather?units=Imperial&APPID=" + api_key 

```

```python
settings = {"units": "metric", "appid": api_key}
```

```python
current_weather_paris = owm.get_current("Paris", **settings)
print(f"Current weather object for Paris: {current_weather_paris}.")
```

```python
print(current_weather_paris['main']['temp'])
```

```python
temp = {}
lat = {}
humid = {}
cloud = {}
wind = {}


for city in cities:
    try:
        current_weather = owm.get_current(city, **settings)
        temp[city] = current_weather['main']['temp']
        humid[city] = current_weather['main']['humidity']
        cloud[city] = current_weather['clouds']['all']
        wind[city] = current_weather['wind']['speed']
        lat[city] = current_weather['coord']['lat']
        print(current_weather['name'],current_weather['id'])
    except:
        print(city, 'N/a')
        
        
w_dict = [temp, lat, humid, cloud, wind]

weather_df = pd.DataFrame(w_dict)

weather_df.head()
```

```python

weather_df = weather_df.T

weather_df.columns = ['Temp','Lat','Humid','Cloud','Wind']


weather_df.head()
```

```python
weather_df.plot(kind='scatter', x='Lat', y='Temp', title='City Temperate vs Latitude 8/19/18')
plt.savefig('Temp.png')
```

```python
weather_df.plot(kind='scatter', x='Lat', y='Humid', title='City Humidity vs Latitude 8/19/18')
plt.savefig('Humid.png')
```

```python
weather_df.plot(kind='scatter', x='Lat', y='Cloud', title='City Cloudiness vs Latitude 8/19/18')
plt.savefig('Cloud.png')
```

```python
weather_df.plot(kind='scatter', x='Lat', y='Wind', title='City Wind vs Latitude 8/19/18')
plt.savefig('Wind.png')
```
